Binary Classification - whether crime occurs given a specific neighborhood and hour of the day.

```python
bpd_df = pd.read_csv('../data/BPD_CRIME_DATA_ENCODED.csv')
```

```python
bpd_df.shape
bpd_df.head(10)
bpd_df = bpd_df.rename(columns={"Unnamed: 0": "Crime_Id"})
```

```python
#add a counter column to make things easier in the next step
ctr = np.array([1 for i in range(bpd_df.shape[0])])
bpd_df['Counter'] = ctr
```

```python
bpd_df.columns
```

We will be predicting the "deadly rate" given a specific neighborhood and a specific day of hour. For now, we use number of Weapon_FIREARM crimes as numerator and number of all crimes as the denominator, so interpretation will be proportion of deadly crimes among all crimes.

```python
uc1_df = bpd_df.groupby(["Hour","Neighborhood"]).agg({"Counter":'sum', "Weapon_FIREARM":"sum"}).reset_index().rename(columns={'Counter':'Total_Crimes', 'Weapon_FIREARM':'Deadly_Crimes'})
uc1_df['Deadly_Rate'] = uc1_df['Deadly_Crimes']/uc1_df['Total_Crimes']
uc1_df
```
