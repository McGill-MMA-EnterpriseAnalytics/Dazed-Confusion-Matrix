# Casual Inference With Binary Targets

This model was used to test inference when for each description category as a binary outcome. However, it was slow to run even when using sampling and the results between categories did not change when using Causal Lift algorithm. Therfore, it is more of a proof of concept and we are using the "CausalResultsMulti" file as the main reference point.

# Setup

```python
import warnings
warnings.filterwarnings("ignore")
%run CausalModel.py
from CausalModel import causal_lift_model
import pandas as pd
pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', None)  # or 199
```
