```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS 

text = 'XGBClassifier Crash Crash KernelDead KernelDead OneVsRest OneVsRest GridSearchCV BlueScreen XGBClassifier LinearSVC DecisionTree RandomForest LightGBMClassifier LightGBMClassifier LightGBMClassifier RidgeClassifier XGBRFClassifier XGBRFClassifier Regularization BernoulliNB BernoulliNB GaussianProcessClassifier MLPClassifier PassiveAggressiveClassifier LinearDiscriminantAnalysis SGDClassifier'

def generate_wordcloud(text): # optionally add: stopwords=STOPWORDS and change the arg below
    wordcloud = WordCloud(font_path='/Library/Fonts/Verdana.ttf',
                          width=800, height=400,
                          relative_scaling = 1.0,
                          background_color="white",
                          stopwords = {'to', 'of'} # set or space-separated string
                          ).generate(text)
    
    fig = plt.figure(1, figsize=(18, 12))
    plt.axis('off')
    plt.imshow(wordcloud)
    plt.axis("off")
    ## Pick One:
    # plt.show()
    plt.savefig("./images/WordCloud.png")

generate_wordcloud(text)
```
